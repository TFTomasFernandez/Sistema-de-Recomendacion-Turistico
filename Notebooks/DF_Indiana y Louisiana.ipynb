```python
import pandas as pd
import json
import os
import glob
```

<h2>Trabajamos los archivos de Indiana <h2>

```python
# Ruta donde est치n almacenados los archivos JSON
path = r'G:/Proyectos Henry/Proyecto Final/Google Maps - Yelp!/Google Maps/reviews-estados/review-Indiana'

# Usamos glob para obtener todos los archivos JSON en el directorio
all_files = glob.glob(os.path.join(path, "*.json"))

# Creamos la lista para almacenar los DataFrames
dfs = []

# Leemos cada archivo JSON y a침adirlo a la lista de DataFrames
for file in all_files:
    df = pd.read_json(file, lines=True)
    dfs.append(df)

# Concatenamos todos los DataFrames en uno solo
df_combined = pd.concat(dfs, ignore_index=True)

df_indiana = df_combined

# Mostramos las primeras filas del DataFrame combinado
df_indiana.head(3)
```

```python
# Eliminamos las columnas 'name', 'pics' y 'resp' del DataFrame combinado
df_indiana = df_indiana.drop(columns=['name', 'pics', 'resp'])

# Mostramos las primeras filas del DataFrame combinado
df_indiana.head(3)
```

```python
# Convertimos microsegundos fecha y agregamos una columna con el valor
df_indiana["date"] = pd.to_datetime(df_indiana["time"], unit='ms')

#Eliminamos la columna time ya que no sirve su contenido
df_indiana = df_indiana.drop(columns=['time'])

# Mostramos las primeras filas del DataFrame combinado
df_indiana.head(3)
```

```python
# Exportamos el DataFrame a un archivo Parquet
df_indiana.to_parquet('ReviewIndiana', engine='pyarrow', index=False)
```
